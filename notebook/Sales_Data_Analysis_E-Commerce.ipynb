# Sales Data Analysis(E- Commerce)

This project is focused on analyzing an E-commerce business. The dataset includes order ID, product information, customer details, quantity, price, and location. The key objectives of the analysis are to determine the most sold products, identify the month with the highest sales volume, and pinpoint the city with the highest sales.
The data is stored in the feather format, a lightweight binary file format designed for efficient data storage and exchange between programming languages, especially for data analysis and machine learning workflows. To work with feather files, the pyarrow package needs to be installed. 
The use of "r" denotes row strings, enabling Pandas to read the file efficiently.
As part of the project, rows with missing data will be dropped, and duplicates have already been eliminated.

This project aims to answer several key questions related to the E-commerce business analysis:

   - What is the best month in terms of sales?
   - Which city has the highest sales volume?
   - Which product generates the highest revenue for the business?
   - What products are most frequently purchased by clients?
   - Which products are commonly sold together?


Project Parts:

   - Data Extraction:
        Extracting the necessary data for analysis from the E-commerce business dataset.
   
   - Data Transformation:
        - Searching for missing data in the dataset.
        - Dropping the rows with missing data to ensure data completeness.
        - Identifying and searching for any duplicate entries in the dataset.
        - Dropping duplicate entries to ensure data accuracy.
        
   - Monthly Sales Analysis:
        Analyzing the sales data on a monthly basis to identify patterns and trends.
        
   - Month Extraction:
        Extracting the month from the available dates in the dataset.
        
   - Data Type Conversion:
        Converting certain data types, if needed, to facilitate analysis and visualization.
        
   - Answering Questions through Data Visualization:
        Utilizing data visualization techniques to answer specific questions related to the E-commerce business.
        

By following these project parts, we can effectively extract and transform the data, analyze monthly sales, extract relevant information such as the month, convert data types if required, and provide answers to specific questions through the visualization of the data. This approach enables a comprehensive analysis of the E-commerce business dataset, leading to valuable insights and informed decision-making.

The unstack() function is used to reshape the hierarchical index created by the groupby() operation into a more tabular format.

In the code most_sold_products_df.groupby(['Month', 'Product']).size().unstack(), the unstack() function is applied after size() to pivot the grouped data. It takes the innermost level of the index (in this case, 'Product') and converts it into column headers, resulting in a DataFrame with 'Month' as the index and 'Product' as separate columns.

```python
pivot=most_sold_products_df.groupby(['Month', 'Product']).size().unstack()
pivot
```
